```python
#cargamos las librerias que utilizamos
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importamos los datos

datos = pd.read_csv('/content/Churn_Modelling.csv')
#eliminamos las columnas que no utilizamos
#del datos['RowNumber']
#del datos['CustomerId']

datos.head()
```

```python
# nos quedamos con el las variables del 3 al 13 en X(eliminando las columnas 0,1 y 2),
# y con la 13 como variable dependiente
X = datos.iloc[:, 3:13].values
y = datos.iloc[:, 13].values

```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(categories='auto'), [1])],   
    remainder='passthrough'                        
)
X = onehotencoder.fit_transform(X)
X = X[:, 1:]
```

```python
#Dividimos los datos en train y test, ademas escalamos los datos

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()

#a単adimos la primera capa
classifier.add(Dense(units = 6, kernel_initializer = "uniform",  
                     activation = "relu", input_dim = 11))
```

```python
#a単adimos la 2 capa

classifier.add(Dense(units = 6, kernel_initializer = "uniform",  activation = "relu"))

```

```python
#a単adimos la 3 capa y finalmente compilamos

classifier.add(Dense(units = 1, kernel_initializer = "uniform",  activation = "sigmoid"))

#compilamos

classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
#ajustamos el RNA al conjunto de entrenamiento

classifier.fit(X_train, y_train,  batch_size = 10, epochs = 100)
```

```python
#finalmente evaluamos el modelo con una matriz de confusion

y_pred  = classifier.predict(X_test)
y_pred = (y_pred>0.5)
y_pred
```

```python
#realizamos una matriz de confusion para ver que tan bueno es el modelo
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
pd.DataFrame(cm)
```

```python
salud = pd.read_csv('/content/Maternal_Health_Risk_Data_Set.csv')
del salud['Unnamed: 0']
salud.head()
```

```python
#nos quedamos en X con las variables, Age,SystolicBP,	DiastolicBP,	BS,	BodyTemp,	HeartRate, mientras que en Y
# nos quedamos con RiskLevel	
X = salud.iloc[:, 0:6].values
y = salud.iloc[:, 6].values
```

```python
#dividimos el datasets en entrenamiento y testing.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
#construimos el RNA

import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()
classifier.add(Dense(units = 12, kernel_initializer = "uniform",  
                     activation = "relu", input_dim = 6))
```

```python
#anadimos la 2 capa
classifier.add(Dense(units = 6, kernel_initializer = "uniform",  activation = "relu"))
```

```python
#a単adimos la capa de salida
classifier.add(Dense(units = 1, kernel_initializer = "uniform",  activation = "sigmoid"))
```

```python
#finalmente compilamos el modelo
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
#ajustamos el conjunto de entrenamiento

classifier.fit(X_train, y_train,  batch_size = 10, epochs = 100)
```

```python
#prediccion de los resultados
y_pred  = classifier.predict(X_test)
y_pred = (y_pred>0.5)
```

```python
#realizamos una matriz de confusion para ver que tan bueno es el modelo
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
confusion = pd.DataFrame(cm)
confusion
```

```python
TN=confusion[0][0]
TP=confusion[1][1]
FN=confusion[0][1]
FP=confusion[1][0]
#calculamos la sensibilidad
sens = TP/(TP+FN)
sens
```

Red Neuronal con Tensorflow 2xx

```python
#cargamos los datos
fashion = pd.read_csv('/content/fashion-mnist_train.csv')
fashion.head()
```

```python
# nos quedamos con el las variables del 3 al 13 en X(eliminando las columnas 0,1 y 2),
# y con la 13 como variable dependiente
X = fashion.iloc[:, 1:785].values
y = fashion.iloc[:, 0].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
#redimensionamos el data
X_train = X_train.reshape(-1, 28*28)
X_train.shape
```

```python
#redimensionamos tambien el testing
#Redimensionamos el conjunto de testing del mismo modo
X_test = X_test.reshape(-1, 28*28)
```

```python
import numpy as np
import datetime
import tensorflow as tf
model = tf.keras.models.Sequential()
```

```python
#agregamos la primera capa
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
```

```python
model.add(tf.keras.layers.Dropout(0.2))
```

```python
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
```

```python
model.summary()
```

```python
#entrenamos el modelo

model.fit(X_train, y_train, epochs=9)
```

```python
#calculamos la prediccion del modelo
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
print("Test accuracy: {}".format(test_accuracy))
```

```python
model_json = model.to_json()
with open("fashion_model.json", "w") as json_file:
    json_file.write(model_json)
```
