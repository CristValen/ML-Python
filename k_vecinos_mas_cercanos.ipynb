```python
import numpy as np
from sklearn.model_selection import  cross_validate, train_test_split
import pandas as pd
from sklearn import preprocessing, neighbors

#trabajamos con datos ya clasificados con cualquier tecnica, y una vez clasificado, posicionamos nuevos datos y aplicmaos K-vecinos.
#Aqui utilizamos un datasets de tumores

tumores = pd.read_csv('/content/tumores.csv')
del tumores['Unnamed: 0']
tumores.head()
```

```python
#comprobamos que no existan NAs

tumores.describe()
```

```python
#Creamos la X e Y
Y = tumores["Class"]
X = tumores[['Adhes','BNucl','Chrom',	'Epith'	,'Mitos'	,'NNucl'	,'Thick'	,'UShap'	,'USize']]
X.head()
```

```python
Y.head
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X.values, Y.values, test_size=0.3)
```

```python
clf = neighbors.KNeighborsClassifier()
```

```python
#agregamos ".values" para evitar errores de nombres o etiquetas
clf.fit(X_train, Y_train)
```

```python
accuracy = clf.score(X_test, Y_test)
accuracy
```

```python
#utilizamos los datos clasificados para clasificar nuevos datos, supongamos que llega un paciente y se hace las pruebas
# y obtiene medidas como 2,4,6,2,1,5,6,3,2

paciente_nuevo = np.array([2,4,6,2,1,5,6,3,2])
paciente_nuevo = paciente_nuevo.reshape(1,-1)

```

```python
predict = clf.predict(paciente_nuevo)
predict
#obtenemos que el paciente tendria un tumor "maligno (0)"
```

```python
#si tuvieramos dos pacientes
dos_pacientes = np.array([[4,2,1,1,1,2,3,2,1], [2,2,4,4,2,2,6,2,4]]).reshape(2,-1)
predict = clf.predict(dos_pacientes)
predict
#en este caso ambos pacientes tendrian tumores "benignos[1,1]"
```
